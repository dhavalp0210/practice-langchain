```python
from multilingual_pdf2text.pdf2text import PDF2Text
from multilingual_pdf2text.models.document_model.document import Document
```

```python
pdf2text=PDF2Text(document=pdf_document)
content=pdf2text.extract()
```

```python
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Ensure all required fields are provided
documents = [
    Document(
        page_content=item["text"],
        metadata={"page_number": item["page_number"]},
        # document_path="example_path"  # Add this field if required
    )
    for item in content
]

text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=300)
chunks = text_splitter.split_documents(documents)

for chunk in chunks:
    print(f"Page: {chunk.metadata['page_number']}, Content: {chunk.page_content}")
```

```python
chunks
```

```python
output_file_path = r"E:\granth-rag\extract-pdf\guj-san-demo-chunks.txt"

with open(output_file_path, 'w', encoding='utf-8') as text_file:
    for chunk in chunks:
        # Retrieve page number from metadata
        page_number = chunk.metadata.get('page_number', 'Unknown')
        content = chunk.page_content
        text_file.write(f"Page Number: {page_number}\n")
        text_file.write(f"Content:\n{content}\n")
        text_file.write("\n" + "-"*80 + "\n\n")  
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain_core.documents import Document

# Initialize the HuggingFaceEmbeddings with the KaLM model
embedding_model = HuggingFaceEmbeddings(model_name="HIT-TMG/KaLM-embedding-multilingual-mini-instruct-v1")
```

```python
from langchain_community.vectorstores.faiss import FAISS, DistanceStrategy

# Initialize the FAISS vector store with the desired distance metric
vector_store = FAISS.from_documents(
    chunks,
    embedding_model,
    distance_strategy=DistanceStrategy.COSINE
)
```
