```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="this is a tutorial on openai embedding"
query_result=embeddings.embed_query(text)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
speech=""
with open("speech.txt") as f:
    speech=f.read()



text_splitter = RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
text = text_splitter.create_documents([speech])
print(text)
```

```python
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(text,embeddings)
db
```

```python
query="Speech recognition is an interdisciplinary subfield of computer science and computational linguistics that develops methodologies and technologies that enable the recognition and translation of spoken language into text by computers. It is also known as automatic speech recognition (ASR), computer speech recognition or speech-to-text (STT). It incorporates knowledge and research in the computer science, linguistics and computer engineering fields. The reverse process is speech synthesis."
retrived_results=db.similarity_search(query)
print(retrived_results)
```
